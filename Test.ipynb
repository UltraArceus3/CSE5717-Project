```python
import cv2 as cv
from glob import glob
import os
import random
from ultralytics import YOLO
```

```python
# read in video paths
videos = glob('vid.mp4')
print(videos)
```

```python
# pick pre-trained model
model_pretrained = YOLO('yolov8n.pt')
```

```python
# read video by index
video = cv.VideoCapture("vid.mp4")

# get video dims
frame_width = int(video.get(3))
frame_height = int(video.get(4))
size = (frame_width, frame_height)

# Define the codec and create VideoWriter object
fourcc = cv.VideoWriter_fourcc(*'DIVX')
out = cv.VideoWriter('./outputs/vid.mp4', fourcc, 20.0, size)

# read frames
ret = True

while ret:
    ret, frame = video.read()

    if ret:
        # detect & track objects
        results = model_pretrained.track(frame, persist=True)

        # plot results
        composed = results[0].plot()

        # save video
        out.write(composed)

out.release()
video.release()


```
